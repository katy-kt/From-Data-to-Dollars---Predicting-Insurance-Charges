```python
# Re-run this cell
# Import required libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

# Loading the insurance dataset
insurance_data_path = 'insurance.csv'
insurance = pd.read_csv(insurance_data_path)
insurance.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression  # 線性迴歸模型
from sklearn.metrics import r2_score as calculate_r2 # 計算 R2 分數

# 1. 讀取歷史資料與驗證資料
insurance = pd.read_csv('insurance.csv')
validation_data = pd.read_csv('validation_dataset.csv')

# 清洗 charges 欄位：移除$字號、逗號，並轉為浮點數
if insurance['charges'].dtype == 'object':
    insurance['charges'] = insurance['charges'].str.replace('$', '', regex=False)
    insurance['charges'] = insurance['charges'].str.replace(',', '', regex=False)
    insurance['charges'] = insurance['charges'].astype(float)

# 處理缺失值：將含有空值 (NaN) 的資料直接刪除
insurance = insurance.dropna()

# 2. 資料前處理：把剩餘的文字（如：性別、區域、抽菸）轉成數字
insurance_cleaned = pd.get_dummies(insurance, drop_first=True)
validation_cleaned = pd.get_dummies(validation_data, drop_first=True)

# 確保驗證資料的欄位順序跟訓練資料完全一樣（除去了目標欄位 charges）
X_train = insurance_cleaned.drop(columns=['charges'])
y_train = insurance_cleaned['charges']

# 對齊欄位，防止兩份資料的文字特徵數量不一致
X_val = validation_cleaned.reindex(columns=X_train.columns, fill_value=0)

# 3. 建立並訓練迴歸模型
model = LinearRegression()
model.fit(X_train, y_train)

# 4. 在訓練集上評估模型，並將 R2 分數存入規定變數 r2_score
y_pred_train = model.predict(X_train)
r2_score = calculate_r2(y_train, y_pred_train)
print(f"你的模型 R2 分數為: {r2_score:.4f} (大於 0.65 即可)")

# 5. 預測新資料的保費
predictions = model.predict(X_val)

# 6. 特殊條件：確保最低基本保費為 1000
predictions = np.where(predictions < 1000, 1000, predictions)

# 7. 將預測結果存回指定的 DataFrame 與欄位
validation_data['predicted_charges'] = predictions
```
